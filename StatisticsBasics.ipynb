Data can be classified into two major types:

Qualitative data describes attributes, labels, or names. It cannot be measured in numbers but can be observed or grouped.

>Definition: Categories that do not have an inherent order or ranking.

>Definition: Numbers where the differences are meaningful, but no true zero exists.

>Examples:


*   Temperature in Celsius or Fahrenheit (0Â°C does not mean no temperature)

>Use: Scientific and economic analysis.

5. Discuss the role of random sampling in making inferences about populations.

Ans: Role of Random Sampling in Inference

Random Sampling is a technique where every individual in a population has an equal chance of being selected.

Why It Matters:

> Ensures representativeness

> Avoids bias

> Makes statistical inference valid

Statistical Inference:

Drawing conclusions about a whole population based on a sample.

Key Concepts:

> Population: Entire group (e.g., all college students)

> Sample: Subset selected randomly

> Inference: Using sample statistics (mean, SD) to estimate population parameters

Example:

> You survey 500 randomly selected students about study habits.

> Based on the responses, you estimate that 60% of all students prefer studying at night.

> Because your sample was random, you can confidently generalize this to the broader population.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

Ans: What is Skewness?

Skewness is a measure of the asymmetry or lack of symmetry in the distribution of data.

> In a perfectly symmetrical distribution, the mean = median = mode.

> If data is not symmetrical, it is said to be skewed.

Types of Skewness:

1. Positive Skew (Right-Skewed)

> Tail extends to the right

> Mean > Median > Mode

> Examples:


*   Income distribution (few people earn a lot more)
*   Test scores where most students failed, but a few scored high


2. Negative Skew (Left-Skewed)

> Tail extends to the left

> Mean < Median < Mode

> Examples:

*   Age at retirement (most people retire around 60â€“65, few retire much earlier)
*  Test scores where most students scored high, few scored low


3. Zero Skew (Symmetrical)

> Mean â‰ˆ Median â‰ˆ Mode

> Example: Heights of adults in a population


Effect on Interpretation
> Skewness impacts central tendency:

*   Mean is pulled toward the tail.
*   Median is a better measure for skewed data.

> Skewness helps in selecting the right model or transformation for analysis.
